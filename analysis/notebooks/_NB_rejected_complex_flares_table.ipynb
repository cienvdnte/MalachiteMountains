```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../data/gaia/21_10_2020_manual_vetting.csv")
```

```python
df = df[df.note == "complex shape"]
df.head().T
```

```python
df3 = df[["TIC", "Sector", "period_h", "tstart"]]
df3.period_h = df3.period_h.round(2)
df3 = df3.rename(index=str, columns={"period_h":r"$P$ [h]",
                                     "tstart":r"$t_{\text{start}}$ [BTJD - 2457000 d]",})
```

```python
stri = df3.to_latex(index=False,escape=False, column_format=f"l|ccc")
stri = stri.replace("\\toprule","\hline\hline")
stri = stri.replace("\\midrule","\hline")
stri = stri.replace("\\bottomrule","\hline\n" )
```

```python
print(stri)
```

```python
PWD = "/home/ekaterina/Documents/002_writing/multiperiod-flares-draft-mnras/tables"

# Write out latex string
with open(f"{PWD}/complex_flares.tex", "w") as f:
    f.write(stri)
```
